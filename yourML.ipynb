```python
def calculate_ideal_sleep_duration(age, activity_level, stress, heartrate):
    sleep_duration = 7.5
    
    if age < 25:
        sleep_duration += 0.5
    elif age > 50:
        sleep_duration -= 0.5
    
    if activity_level == 'very_active':
        sleep_duration += 0.5
    elif activity_level == 'sedentary':
        sleep_duration -= 0.5
    
    if stress == 'high':
        sleep_duration += 0.5
    elif stress == 'low':
        sleep_duration -= 0.5
    
    if heartrate > 80:
        sleep_duration += 0.5
    
    return min(max(sleep_duration, 6), 9)  
```

```python
# Assuming df is your existing DataFrame
ideal_sleep_duration = [
    calculate_ideal_sleep_duration(df.at[i, 'age'], df.at[i, 'activity_level'], df.at[i, 'stress'], df.at[i, 'heartrate']) 
    for i in range(len(df))
]


df['ideal_sleep_duration'] = ideal_sleep_duration


```

```python
features = df.drop(columns=['daily_caloric_intake', 'daily_step_goal', 'ideal_sleep_duration'])
targets = df[['daily_caloric_intake', 'daily_step_goal', 'ideal_sleep_duration']]
```

```python
features.fillna(features.mode().iloc[0], inplace=True)
targets.fillna(targets.mode().iloc[0], inplace=True)
```

```python
import pandas as pd


def remove_outliers(df, columns):
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
    return df


numerical_cols = ['height', 'weight', 'heartrate', 'age', 'systolic_bp', 'diastolic_bp', 'daily_step_goal', 'daily_caloric_intake','ideal_sleep_duration']
df = remove_outliers(df, numerical_cols)
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, targets, test_size=0.2, random_state=42)
```

```python
numeric_features = ['age', 'height', 'weight', 'heartrate', 'systolic_bp', 'diastolic_bp']
categorical_features = ['diet', 'gender','alcohol_smoke_both','stress']
ordinal_encoder = ['activity_level']
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder, OrdinalEncoder
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import mean_squared_error
import joblib
```

```python
numeric_features = ['age', 'height', 'weight', 'heartrate', 'systolic_bp', 'diastolic_bp']
categorical_features = ['diet', 'gender','alcohol_smoke_both','stress']
ordinal_encoder = ['activity_level']
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_features),
        ('cat', OneHotEncoder(), categorical_features),
        ('ord', OrdinalEncoder(), ordinal_encoder),
    ])
```

```python
model = RandomForestRegressor(n_estimators=100, random_state=42)
```

```python
pipeline = Pipeline(steps=[('preprocessor', preprocessor),
                           ('model', model)])
```

```python
pipeline.fit(X_train, y_train)
y_pred = pipeline.predict(X_test)
mse = mean_squared_error(y_test, y_pred, multioutput='raw_values')
print(f'Mean Squared Errors: {mse}')
```

```python
param_grid = {
    'model__n_estimators': [50, 100, 200],
    'model__max_depth': [None, 10, 20, 30]
}
```

```python
grid_search = GridSearchCV(pipeline, param_grid, cv=5, scoring='neg_mean_squared_error', verbose=2)
grid_search.fit(X_train, y_train)
best_model = grid_search.best_estimator_
joblib.dump(best_model, 'mymodel.pkl')
```
